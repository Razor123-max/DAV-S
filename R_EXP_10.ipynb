```r
library(ggplot2)
library(dplyr)
```

```r
df <- read.csv("Housing.csv")
df_subset <- head(df, 10)
df_subset
```

```r
ggplot(df_subset, aes(x = factor(bedrooms), y = price)) +
  geom_bar(stat = "identity", fill = "steelblue") +
  labs(title = "House Prices for First 10 Rows", x = "Number of Bedrooms", y = "Price") +
theme_minimal()
```

```r

ggplot(df_subset, aes(x = area, y = price)) +
  geom_point(color = "red", size = 3) +
  labs(title = "Price vs Area (First 10 Rows)", x = "Area (sq ft)", y = "Price") +
  theme_minimal()
```

```r
library(dplyr)
furnish_counts <- df_subset %>%
  count(furnishingstatus)


ggplot(furnish_counts, aes(x = "", y = n, fill = furnishingstatus)) +
  geom_bar(stat = "identity", width = 1) +
  coord_polar("y") +
  labs(title = "Furnishing Status (First 10 Rows)", fill = "Furnishing Status") +
  theme_void()
```
