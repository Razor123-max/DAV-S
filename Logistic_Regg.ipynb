```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

# Step 1: Input Data
# X = hours studied
X = np.array([[1], [2], [3], [4], [5]])
# y = 0 = fail, 1 = pass
y = np.array([0, 0, 0, 1, 1])

# Step 2: Create & Train Model
model = LogisticRegression()
model.fit(X, y)

# Step 3: Visualization
# Create a range of values from 0 to 6 hours
x_test = np.linspace(0, 6, 300).reshape(-1, 1)
# Predict probabilities for class 1 (pass)
y_prob = model.predict_proba(x_test)[:, 1]

# Step 4: Plot
plt.figure(figsize=(8, 5))
plt.scatter(X, y, color='red', label='Actual Data')
plt.plot(x_test, y_prob, color='blue', label='Logistic Curve (Pass Probability)')
plt.axhline(0.5, color='gray', linestyle='--', label='Decision Boundary (0.5)')

plt.xlabel('Hours Studied')
plt.ylabel('Probability of Passing')
plt.title('Logistic Regression - Student Pass Prediction')
plt.legend()
plt.grid(True)
plt.show()
```
