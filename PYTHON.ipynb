```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data =  pd.read_csv("Housing.csv")
data
```

```python
data.info()
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x = "bedrooms", y = "price", data = data)
plt.title("Average House Price by Number of Bedrooms")
plt.xlabel("Number of Bedrooms")
plt.ylabel("Average Price")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x="area", y="price", data=data, color="blue")
plt.title("House Price vs. Area")
plt.xlabel("Area (sq ft)")
plt.ylabel("Price")
plt.show()
```

```python
furnish_counts = data["furnishingstatus"].value_counts()
furnish_counts
```

```python
plt.figure(figsize=(7, 7))
plt.pie(furnish_counts, labels=furnish_counts.index, autopct="%1.1f%%", colors=["blue", "green", "orange"])
plt.title("Distribution of Houses by Furnishing Status")
plt.show()
```
