```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

iris = load_iris()

df = pd.DataFrame(iris.data, columns=iris.feature_names)

df['target'] = iris.target
df['target_name'] = df['target'].apply(lambda x: iris.target_names[x])


df['variety'] = iris.target
df['variety'] = df['variety'].apply(lambda i: iris.target_names[i])
df
```

```python
df.head()
```

```python
df.info()
```

```python
plt.figure(figsize = (10, 8))
sns.pairplot(df, hue = 'variety', palette = 'viridis')
```

```python
plt.scatter(df['sepal length (cm)'], df['sepal width (cm)'], c = df['target'])
```

```python
sns.countplot(x='variety', data=df)
plt.title("Count of Each Variety")
plt.ylabel("Count")
plt.xlabel("Iris Variety")
plt.show()
```

```python
sns.boxplot(x='variety', y='petal length (cm)', data=df)
plt.title("Petal Length Comparison Across Varieties")
plt.show()
```

```python
sns.histplot(df['petal length (cm)'], kde = True, color='r')
```
